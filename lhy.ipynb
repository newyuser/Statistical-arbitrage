```python
#初始化

import qlib
from qlib.config import REG_CN

provider_uri="./qlib_data/my_data"
qlib.init(provider_uri=provider_uri,region=REG_CN)

from qlib.data import D
# trade_date = D.calendar(start_time='2020-12-01', end_time='2020-12-31', freq='day')
instruments = D.instruments(market='all')
# stock_list = D.list_instruments(instruments=instruments,
#                                 start_time='2020-12-01',
#                                 end_time='2020-12-31',
#                                 as_list=True)
```

```python
#获取成交价并格式化

fields = ['$close','$open']
df_close = D.features(instruments, fields, start_time='2020-12-01', end_time='2020-12-31', freq='day')
print(df_close.columns)

group = df_close.groupby('instrument')

import pandas as pd
dfs = []
for g in list(group):
    item = g[1]
    item.rename(columns={'$close':g[0]}, inplace = True)
    item.index = item.index.get_level_values('datetime')
    dfs.append(item)
prices = pd.concat(dfs, axis=1)
prices
```
